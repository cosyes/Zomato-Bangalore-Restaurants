# Zomato
Zomato is an Indian multinational restaurant aggregator and food delivery company founded by Deepinder Goyal and
Pankaj Chaddah in 2008. Zomato provides information, menus and user-reviews of restaurants as well as food delivery 
options from partner restaurants in select cities.

Bengaluru (also called Bangalore) is the capital of India's southern Karnataka state. The center of India's high-tech industry,

### The dataset contains a lot of information about the restaurants in Bangalore.

**This is description of dataset.**

•	URL: contains the URL of the restaurant on the zomato website

•	address=contains the address of the restaurant in Bengaluru

•	name: contains the name of the restaurant

•	online_order=whether online ordering is available in the restaurant or not

•	book_table=table book option available or not

•	rate=contains the overall rating of the restaurant out of 5

•	votes=contains total number of rating for the restaurant as of the above-mentioned date

•	phone=contains the phone number of the restaurant

•	location=contains the neighborhood in which the restaurant is located

•	rest_type=restaurant type

•	dish_liked=dishes people liked in the restaurant

•	cuisines=food styles, separated by a comma

•	approx_cost(for two people)=contains the approximate cost for a meal for two people

•	reviews_list=list of tuples containing reviews for the restaurant, each tuple consists of two values

#### First let's remove the column which seems to be the least important in this analysis.

```python
#removing all data that seems less useful to prevent data lost during cleaning process
zomato_df.drop(['url','address','phone','menu_item','dish_liked'], inplace=True, axis=1) 
```

#### Changing the name of some columns to access their data propery

```python
#Change the names of some columns to access their data properly without getting any errors due to improper column names.
zomato_df.rename(columns = {'approx_cost(for two people)':'approx_cost_for_two_people',
                           'listed_in(type)':'listed_in_type','listed_in(city)':'listed_in_city'}, inplace = True)
zomato_df
```

```python
#split rate as delimiter "/"  and accepting the first indexing[0] value only as rating_out_of_5
#Repalce commas "," from approx_cost_for_two_people and then converting it to float type

zomato_df['rating_out_of_5'] = ((zomato_df['rate'].str.split('/',expand=True))[0]) 
zomato_df['approx_cost_for_two_people'] = zomato_df['approx_cost_for_two_people'].str.replace(',', '')
zomato_df['approx_cost_for_two_people'] = zomato_df['approx_cost_for_two_people'].astype(float)
```

```python
zomato_df.drop(['rate'], inplace=True, axis=1)    # removing/droping rate column 
zomato_df
```

```python
zomato_df.drop_duplicates(keep=False,inplace=True) # removing all the duplicates datas from dataset
zomato_df
```

```python
zomato_df.dropna(inplace = True) #removing all NaN and empty datas along with whole row
zomato_df
```

## Q. What percentage of restaurants are not taking online orders?

```python
Online_order_option = zomato_df.online_order.value_counts() #counting the number of yes and no in zomato_df.online_order 
sns.set(font_scale = 1.2)
labels = 'YES','NO'
sizes = Online_order_option
explode = (0,0.1)  # only "explode" the 2nd slice
color = ['#FF6347','#FF4040']
fig1, ax1 = plt.subplots()
plt.title("Online order option")
ax1.pie(sizes, explode=explode, labels=labels,colors=color ,autopct='%1.1f%%',
        shadow=True, startangle=65)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show();
```

**A high number (35.3%) of restaurants are not taking online orders.** 

## Q. What percentage of restaurants have the option of table booking?

```python
Table_booking_option = zomato_df.book_table.value_counts()  #counting the number of yes and no in zomato_df.book_table
sns.set(font_scale = 1.2)
labels = 'YES','NO'
sizes = Table_booking_option
explode = (0,0.1) 
color = ['#FF6347','#FF4040']
fig1, ax1 = plt.subplots()
plt.title("Table booking option")
ax1.pie(sizes, explode=explode, labels=labels,colors=color,autopct='%1.1f%%',
        shadow=True, startangle=45)
ax1.axis('equal') 
plt.show();
```
